```python
import openai
from google.colab import userdata

openai.api_key = userdata.get('OPENAI_API_KEY')

query = "프랑스의 수도를 알려줘."

messages = [{
 "role": "user",
 "content": query
 }]
model = "gpt-3.5-turbo-0125"

response = openai.ChatCompletion.create(
 model=model,
 messages=messages
)
answer = response.choices[0].message.content
print(answer)

messages = [{
 "role": "system",
 "content": "You are a helpful assistant."
 }, {
 "role": "user",
 "content": query
 }]
model = "gpt-3.5-turbo-0125"

response = openai.ChatCompletion.create(
 model=model,
 messages=messages
)
answer = response.choices[0].message.content
print(answer)

messages = [{
 "role": "system",
 "content": "You are a humorous assistant."
 }, {
 "role": "user",
 "content": query
 }]
model = "gpt-3.5-turbo-0125"

response = openai.ChatCompletion.create(
 model=model,
 messages=messages
)
answer = response.choices[0].message.content
print(answer)

messages = [{
 "role": "system",
 "content": "You are a sarcasm assistant."
 }, {
 "role": "user",
 "content": query
 }]
model = "gpt-3.5-turbo-0125"

response = openai.ChatCompletion.create(
 model=model,
 messages=messages
)
answer = response.choices[0].message.content
print(answer)

messages = [{
 "role": "system",
 "content": "You are a pessimistic assistant."
 }, {
 "role": "user",
 "content": query
 }]
model = "gpt-3.5-turbo-0125"

response = openai.ChatCompletion.create(
 model=model,
 messages=messages
)
answer = response.choices[0].message.content
print(answer)
```
