```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

fine_tuned_model_id = "ft:gpt-3.5-turbo-0125:personal:squidward2:CllVQPg7"

questions = [
    "징징아, 나랑 같이 해파리 사냥 갈래?",
    "오늘따라 클라리넷 소리가 정말 좋은걸.",
    "사장님이 월급을 올려주신대! 정말 신나지 않아?",
    "징징아, 행복이란 도대체 뭘까?",
    "너희 집 바로 옆으로 이사 가도 될까?",
    "징징아, 너 뚱이랑 친구지?",
    "오늘 저녁 메뉴 추천 좀 해줘. 맛있는 걸로!",
    "나 이번에 로또 1등 당첨됐어! 너한테만 알려주는 거야.",
    "징징아, 혹시 춤출 줄 알아? 춤 보여줘!",
    "이 세상에서 제일 싫은 게 뭐야?"
]

print(f"--- Fine-tuned Model Test ({fine_tuned_model_id}) ---\n")

for q in questions:
    completion = client.chat.completions.create(
        model=fine_tuned_model_id,
        messages=[
            {"role": "system", "content": "스폰지밥의 징징이 말투로 염세적이고 짜증 섞인 태도로 답장하시오."},
            {"role": "user", "content": q}
        ]
    )

    answer = completion.choices[0].message.content
    print(f"User: {q}")
    print(f"Squidward: {answer}")
    print("-" * 50)
```
